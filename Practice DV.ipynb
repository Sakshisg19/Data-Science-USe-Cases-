```python
city_hotel=data[data.hotel=='City Hotel']

resort_hotel=data[data.hotel=='Resort Hotel']
```

```python
# filter city hotel and resort hotel
data.value_counts('hotel')
```

```python
# Cancelletion number of both the hotels
can1=city_hotel.value_counts('is_canceled')/len(city_hotel)*100
can2=resort_hotel.value_counts('is_canceled')/len(resort_hotel)*100

can1,can2
```

```python
# Cancelletion according to year
pd.crosstab(data.arrival_date_year,data.is_canceled)
```

```python
#
hotels={}

for hotel in data['hotel'].value_counts().keys():
  h1=data[data['hotel']==hotel]
  hotels[hotel]=round(h1['is_canceled'].sum()/len(h1['is_canceled']),2)

  print(hotels)



if list(hotels.values())[0]>list(hotels.values())[0]:
  print(f'{list(hotels.keys())[0]} has higher cancelletion')
else:print(f'{list(hotels.keys())[1]} has higher cancelletion')
```

```python
# Cancelletion according to years
year={}

for i in city_hotel['arrival_date_year'].value_counts().keys():
  h2=city_hotel[city_hotel['arrival_date_year']==i]
  year[i]=round(h2['is_canceled'].sum()/len(h2['is_canceled']),2)

  print('city_hotel:',year)


for j in resort_hotel['arrival_date_year'].value_counts().keys():
  h3=resort_hotel[resort_hotel['arrival_date_year']==j]
  year[j]=round(h3['is_canceled'].sum()/len(h2['is_canceled']),2)

  print('resort_hotel:',year)

```
