```python
from numpy import array
```

```python
array1=array([[1,2,3,4],[5,6,7,8]])
array1
```

```python
for i in array1:
  print(i)
```

```python
for i in array1:
  for j in i:
    print(j,end=" ")
    print()
```

```python
for i in array1:
  for j in i:
    if(j%3==0):
      print(j)
```
