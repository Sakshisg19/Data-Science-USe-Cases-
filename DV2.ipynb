```python
#data['island'].value_counts().plot(kind='bar')

data.groupby(['island','species']).count()[['bill_length_mm']].unstack().plot(kind='bar',stacked='True',table=True)
plt.xlabel('')
plt.xticks([])
plt.tight_layout()
plt.show()
```
