```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=sns.load_dataset('planets')
data
```

```python
data.duplicated(keep=False)
```

```python
data_columns=data.loc[:,data.columns !='orbital_period']

data_duplicate=data_columns[data_columns.duplicated(keep='first')]
data_duplicate
```

```python
print(data.shape)

data_drop_duplicate=data.drop_duplicates()
print(data_drop_duplicate.shape)
```

```python
print(data.shape)

data_dup_by_col=data_drop_duplicate[data_drop_duplicate.duplicated(['distance','year'],keep=False)]
data_dup_by_col
```

```python
print(data.shape)
```

```python
data.drop(index=range(2),inplace=True)
print(data.shape)
```

```python
data.drop(index=[2,4],inplace=True)
print(data.shape)
data.head()
```

```python
data.drop('mass',axis=1,inplace=True)
data.head(3)
```

```python
data=data.replace(to_replace='Radial Velocity',value='Rad Vel')
data.head(3)
```

```python
data.loc[data['distance']>200,"number"]="high"
data.loc[data['distance']<200,"number"]="low"
data.tail()
```

```python
data['number']=np.where((data['distance']>200),'high',data['number'])
data['number']=np.where((data['distance']<200),'low',data['number'])
data.tail()
```

```python
grouper=data.groupby(['year'])

for year,year_ in grouper:
  print(year)
  print(year_.head(1))
```
