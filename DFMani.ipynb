### **Concatenation **

# Joins

```python
# Create some dummy data

df1=pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
    },index=['K0','K1','K2','K3'])

df2=pd.DataFrame({
    'C':['C0','C1','C2'],
    'D':['D0','D1','D2']
    },index=['K0','K1','K4'])

df3=pd.DataFrame({
    'E':['E0','E1','E2','E3'],
    'F':['F0','F1','F2','F3']
    },index=['K0','K1','K5','K6'])

#Inner Join
inner_join=df1.join(df2,how='inner')
print("\nInner Join:")
print(inner_join)

#Outer Join
outer_join=df1.join(df2,how='outer')
print("\nOuter Join:")
print(outer_join)

#Left Join
left_join=df1.join(df3,how='left')
print("\nLeft Join:")
print(left_join)

#Right Join
right_join=df1.join(df3,how='right')
print("\nRight Join:")
print(right_join)
```
