### Line chart


```python
import numpy as np

x=np.arange(1,13)
y=np.linspace(1,10,20)*2.5

plt.figure(figsize=(3,2))

plt.plot(x,y[1:13],color='red',label='First_line')
plt.plot(x,y[1:13]/12,color='green',label='Second_line')

plt.title('Line Chart',color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0,5) # use for starting and ending limit of x axis
plt.ylim(0,15) ## use for starting and ending limit of y axis
plt.xticks([1,1.5,2,2.5,3,3.5,4,4.5,5]) # used to give points(eg. 1,1.5,....) at x-axis
plt.legend()
plt.grid()
plt.show()      # used for neat representation(not unuseful info except graph )
```

```python
import yfinance as yf
```

```python
eur=yf.Ticker("EURUSD=X")
data=eur.history()

plt.figure(figsize=(13,3))
plt.plot(data.Close)

plt.title('Stock Chart',color='Green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
from matplotlib import colors

color=list(colors.cnames)
eur=yf.Ticker("EURUSD=X")
usd_jpy=yf.Ticker("JPY=X")
jpy_data=usd_jpy.history()

data=eur.history()

plt.figure(figsize=(13,3))
plt.title("EUR/USD",color=np.random.choice(color))
plt.plot(data.Close,color=np.random.choice(color))
plt.xlabel('Date')
plt.ylabel('Close price')
plt.grid()
plt.show()

plt.figure(figsize=(13,3))
plt.plot(jpy_data.Close,color=np.random.choice(color))
plt.xlabel('Date')
plt.ylabel('Close price')
plt.grid()
plt.show()
```
