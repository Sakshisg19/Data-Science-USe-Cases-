```python
c=1
for column in data2.columns:
  t = data2[columns]
  skew=t.skew()
  plt.subplot(5,2,c)
  plt.hist(t,label=skew)
  c+=1

  plt.figure(figsize=(5,3))
  plt.legend()
  plt.suptitle("Skweness")
  plt.title(column)
  plt.tight_layout()
```
