```python
# add 100 items in this list,
# if the number is even add the number
# if the number is multiple of 3 add the square
# otherwise add null value and calculate the null %

x=[]
for i in range(1,101):
  if i%2==0:
   x.append(i)
  elif i%3==0:
   x.append(i**2)
  else:
    x.append(np.nan)
(pd.Series(x).isnull().sum()/len(x))*100
```

```python
# Write a Pandas program to add ,subtract,multiple and divide two Pandas Series.
# Sample Series:[2,4,6,8,10],[1,3,5,7,9]

sample=([2,4,6,8,10])
sample2=([1,3,5,7,9])
serie1=pd.Series(sample)
serie2=pd.Series(sample2)

add=serie1+serie2
sub=serie1-serie2
multi=serie1*serie2
div=serie1/serie2

print(add,sub,multi,div)
```

```python
# To compare the elements of two pandas

sample3=([2,4,6,8,10])
sample4=([1,3,5,7,9])
serie1=pd.Series(sample3)
serie2=pd.Series(sample4)

#condition
for i,j in zip(serie1,serie2):
  print(i,j)
if i>j:
    print("Serie1 is greater")
elif j>i:
    print("Serie2 is greater")
else:
    print("Equal")
```
